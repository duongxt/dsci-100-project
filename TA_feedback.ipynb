Xuan's feedback

Data description needs a clearer, more structured summary.
You describe variables informally, but the assignment asks for a full descriptive summary (number of rows/columns, types, meanings, missingness) in a bullet list or table for both players and sessions. Right now, sessions.csv is only described implicitly through code. Also, avoid vague statements like “data is kind of neat” and instead state concretely what is tidy and what is not.

Be careful about leakage and feature choice.
For Question 1, the stakeholders care about predicting subscription from information available at or near signup. Features like actual_hours and sessions come from future behaviour and would leak information if you pretend to predict subscription “from scratch.” It’s fine for EDA, but your methods section should clarify which variables will be allowed as predictors in the final model.

EDA should eventually be done on the training set only.
For this planning stage it’s okay, but you should explicitly note that in the final project you’ll split first (e.g., 80/20) and restrict EDA and feature engineering decisions to the training data to avoid overfitting.

KNN justification has some misconceptions.

KNN absolutely does come with important assumptions: it relies on a meaningful distance metric, requires standardized numeric features, and is sensitive to outliers and class imbalance.

You mention difficulty incorporating categorical variables but don’t say how you’ll handle them. This is crucial for KNN.

For class imbalance, just saying “increase the weight” is vague; you should mention specific strategies

Methods & plan need more detail and structure.
You outline KNN + GridSearch, but you should also:

specify which metrics matter most for the stakeholders (e.g., recall for subscribers vs. overall accuracy),

mention that StandardScaler will be fit on the training data only inside a pipeline, then applied to the test data.
